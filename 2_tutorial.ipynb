####UNION and UNION BY NAME

It is used to merge the two tables and If you use union it wont check the data, it will merge the data

```python
data1 = [('1','rak'),
         ('2','sak')]
schema1 = 'id STRING, name STRING'

df1 = spark.createDataFrame(data = data1, schema = schema1)
df1.display()
```

```python
data2 = [('3','mahesh'),
         ('4','suresh'),
         ('5','ramesh')]
schema2 = 'id STRING, name STRING'

df2 = spark.createDataFrame(data = data2, schema = schema2)
df2.display()
```

```python
df1.union(df2).display()
```

```python
data3 = [('rak','1'),
         ('sak','2'),
         ('sak','2')]
schema1 = 'name STRING, id STRING'

df3 = spark.createDataFrame(data = data3, schema = schema1)
df3.display()
```

```python
df2.union(df3).display()
```

If you use UNION BY NAME it will check the data and merge with correct column

```python
df2.unionByName(df3).display()
```

Both Union and UnionAll is same

```python
df2.unionAll(df3).display()
```

To remove duplicates while merging the data use distinct function with union

```python
df2.union(df3).distinct().display()
```

```python
df3.select('id',lower('name').alias("lower_name")).display()
```


####Date Functions

It is used to manage the date columns in table CURRENT_DATE(), DATE_ADD(), DATE_SUB()


#####CURRENT_DATE()

```python
df = df.withColumn('curr_date',current_date())
df.display()
```

#####DATE_ADD()

It is used to add days in date-time column


```python
df.display()
```

```python
df = df.withColumn('Week_after',date_add('curr_date',7))
df.display()
```

####DATE_SUB()

It is used to minus days in the date-time column

```python
df = df.withColumn('week_before',date_sub('curr_date',7))

df.display()
```

```python
df.display()
```

####DATE_DIFF() 

It is used to check the date difference between 2 columns

```python
df = df.withColumn('diff_btw_week_before_week_after',date_diff('Week_after','week_before'))

df.display()
```

####DATE_FORMAT()

It is used to change the format of the date column

```python
from pyspark.sql.functions import date_format, col
```

```python
# spark.conf.set("spark.sql.legacy.timeParserPolicy", "LEGACY")
df = df.withColumn('curr_date', date_format('curr_date', 'dd-MM-yyyy'))
df.display()
```

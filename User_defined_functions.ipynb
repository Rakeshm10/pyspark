####USER_DEFINED_FUNCTIONS

In real world using udf is less because of performance issue

Step 1 : First you need to create a UDF

```python
def my_func(x):
    return x*x
```

Step 2: Convert into pyspark UDF

```python
my_udf = udf(my_func)
```

Step 3: Use in your pyspark code

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv")\
    .options(header='true', inferSchema='true')\
            .load("/Volumes/pyspark/source/ff_source/BigMart Sales.csv")
df.display()
```

```python
df.withColumn('new_mrp', round(my_udf('Item_MRP'))).select('Item_MRP','new_mrp').display()
```
